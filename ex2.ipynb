```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ex1
sample_list = []
for i in range(50000):
    x = np.random.normal()
    u = np.random.uniform()
    if -0.8<=x and x<=0.8:
        y = 1
    else:
        y = 0
    if u<=y:
        sample_list.append(x)

plt.hist(sample_list, bins=50, density=True, rwidth=0.8, color='r', alpha=0.5)
```

```python
# ex3
lam = 0.5
u = np.random.uniform(size=50000)
x1 = -(1/lam) * np.log(1 - u)
x2 = np.random.uniform(0, 2*np.pi, 50000)
y1 = x1**0.5*np.cos(x2)

w = np.array([0.1, 0.2, 0.3, 0.2, 0.2])
mu = np.array([-2, -1, 0, 1, 2])
std = np.array([0.5, 0.1, 0.5, 0.2, 0.5])

cdf = np.cumsum(w)
samples = []
for i in range(50000):
    u = np.random.uniform(0, 1)
    for k in range(len(cdf)):
        if cdf[k] > u:
            samples.append(y1[i]*std[k]+mu[k])
            break
plt.hist(samples, bins=100, density=True, rwidth=0.8, color='r', alpha=0.5)
```

```python
# ex3
lam = 0.5
u = np.random.uniform(size=50000)
x1 = -(1/lam) * np.log(1 - u)
x2 = np.random.uniform(0, 2*np.pi, 50000)
y1 = x1**0.5*np.cos(x2)

w = np.array([0.1, 0.2, 0.3, 0.2, 0.2])
mu = np.array([-2, -1, 0, 1, 2])
std = np.array([0.5, 0.1, 0.5, 0.2, 0.5])

cdf = np.cumsum(w)
samples = []
for i in range(50000):
    u = np.random.uniform(0, 1)
    for k in range(len(cdf)):
        if cdf[k] > u:
            x = y1[i]*std[k]+mu[k]
            if mu[k]-0.1<=x<=mu[k]+0.1:
                samples.append(x)
            break
plt.hist(samples, bins=100, density=True, rwidth=0.8, color='r', alpha=0.5)
```

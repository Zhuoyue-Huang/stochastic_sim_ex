```python
# ex3
lam = 0.5
u = np.random.uniform(size=50000)
x1 = -(1/lam) * np.log(1 - u)
x2 = np.random.uniform(0, 2*np.pi, 50000)
y1 = x1**0.5*np.cos(x2)

w = np.array([0.1, 0.2, 0.3, 0.2, 0.2])
mu = np.array([-2, -1, 0, 1, 2])
std = np.array([0.5, 0.1, 0.5, 0.2, 0.5])

cdf = np.cumsum(w)
samples = []
for i in range(50000):
    u = np.random.uniform(0, 1)
    for k in range(len(cdf)):
        if cdf[k] > u:
            x = y1[i]*std[k]+mu[k]
            if mu[k]-0.1<=x<=mu[k]+0.1:
                samples.append(x)
            break
plt.hist(samples, bins=100, density=True, rwidth=0.8, color='r', alpha=0.5)
```

```python
# ex3 sol
def discrete(s, w): # draws a single sample from a discrete distribution defined on s with probabilities w
    cw = np.cumsum(w)
    sample = []

    u = np.random.uniform(0, 1)

    for k in range(len(cw)):
        if cw[k] > u:
            sample = s[k]
            break
    return sample

s = np.array([0, 1, 2, 3, 4]) # support of the discrete distribution (indices)
w = np.array([0.1, 0.2, 0.3, 0.2, 0.2]) # weights of the discrete distribution (probabilities)
mu = np.array([-2, -1, 0, 1, 2]) # mean of the Gaussian components
sigma = np.array([0.5, 0.1, 0.5, 0.2, 0.5]) # standard deviation of the Gaussian components

N = 10000 # number of samples to draw
x = np.zeros(N) # initialize the array to store the samples

for i in range(N):
    samp = discrete(s, w) # sample from the discrete distribution
    x[i] = np.random.normal(mu[samp], sigma[samp], 1) # sample from the Gaussian with the sampled index

plt.hist(x, bins=100 , density=True , rwidth=0.8, color='r', alpha=0.5)
plt.show ()

def mixture_of_gauss (xx , mu , sigma , w):
    yy = np.zeros(len(xx))
    for i in range(len(mu)):
        yy += w[i] * np.exp(-0.5 * (xx - mu[i])**2 / sigma[i]**2) / np.sqrt(2 * np.pi * sigma[i]**2)
    return yy

xx = np.linspace(-4, 4, 1000)
yy = mixture_of_gauss (xx , mu , sigma , w)

plt.hist(x, bins=100 , density=True , rwidth=0.8, color='r', alpha=0.5)
plt.plot(xx , yy , 'k-')
plt.show ()
```
